```python
Q_table = np.zeros(shape = (6,6))


game =  RoomGame()
result , state =  game.reset()
gamma = 0.80
alpha = 0.1
epsilon = 0.5 #explore(random action)



max_run = 6
episodes = 1000
rewards = []
for i in range(episodes):
    for _ in range(max_run):
        
        a = np.random.uniform(0,1)
        if a < epsilon:
            action = np.random.randint(0,5)
        else : 
            action = np.argmax(Q_table[state])
        

        # print(state)
        result, new_state , reward = game.self_run(action)
        # rewards.append(reward)
        
        #Bellman Equation
        Q_table[state][action] += alpha*(reward + gamma*(max(Q_table[new_state])) - Q_table[state][action])
        state = new_state
        
        
        if result:
            # print("Reached")
            result , state = game.reset()
            break
```
